```python
import pandas as pd

# Create a DataFrame from the data
data = {
    'Feature 1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
    'Feature 2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
    'Feature 3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
    'Feature 4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
    'Target (Y)': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot each feature against the target variable
axs[0, 0].scatter(df['Feature 1'], df['Target (Y)'])
axs[0, 0].set_title('Feature 1 vs Target')

axs[0, 1].scatter(df['Feature 2'], df['Target (Y)'])
axs[0, 1].set_title('Feature 2 vs Target')

axs[1, 0].scatter(df['Feature 3'], df['Target (Y)'])
axs[1, 0].set_title('Feature 3 vs Target')

axs[1, 1].scatter(df['Feature 4'], df['Target (Y)'])
axs[1, 1].set_title('Feature 4 vs Target')

# Layout so plots do not overlap
fig.tight_layout()

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Define the features and target variable
X = df[['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']]
y = df['Target (Y)']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the coefficients of the best fit line
coefficients = model.coef_
intercept = model.intercept_

print('Coefficients:', coefficients)
print('Intercept:', intercept)
```
